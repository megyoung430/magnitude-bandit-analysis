```python
import random
import itertools

# Your elements
elements = ["A1", "B1", "C1", "A2", "C2", "A3", "B3", "C3"]

# Existing history
problems = {
    7: ['A1','C1','C3'],
    8: ['C2','A3','A1'],
    9: ['B1','A2','C3'],
    10: ['A2','B3','C2'],
    11: ['A1','B1','C1'],
    12: ['B3','C3','A2'],
    13: ['C1','C2','A1'],
    14: ['B3','A3','B1'],
    15: ['A2','C1','A3'],
    16: ['B3','A1','A2'],
    17: ['B1','A3','C3'],
    18: ['A1','C1','B3'],
    19: ['C1','A2','B3'],
    20: ['C1','C2','C3'],
    21: ['A1','A2','C3'],
    22: ['A1','A3','B3'],
}

# Convert history to a set of sorted tuples (order-independent)
used = {tuple(sorted(v)) for v in problems.values()}

# Generate all possible combinations of 3 elements
all_combos = list(itertools.combinations(elements, 3))

# Filter out already used combinations
available = [combo for combo in all_combos if combo not in used]

def generate_new_problem(problem_dict):
    global available
    
    if not available:
        raise ValueError("No more unique combinations left!")
    
    new_combo = random.choice(available)
    available.remove(new_combo)
    
    next_key = max(problem_dict.keys()) + 1
    problem_dict[next_key] = list(new_combo)
    
    return next_key, new_combo

# Example: generate 5 new problems
for _ in range(5):
    k, combo = generate_new_problem(problems)
    print(k, combo)
```

```python
import random
from collections import Counter

towers = ["A1", "A2", "A3", "B1", "B3", "C1", "C2", "C3"]

def valid_schedule(problems, towers, towers_per_problem=3):
    first6_counts = Counter(t for p in problems[:6] for t in p)
    total_counts = Counter(t for p in problems for t in p)

    min_total_count = (len(problems) * towers_per_problem) // len(towers)

    return (
        all(len(p) == towers_per_problem for p in problems)
        and all(len(set(p)) == towers_per_problem for p in problems)
        and len(set(tuple(sorted(p)) for p in problems)) == len(problems)
        and all(first6_counts[t] >= 2 for t in towers)
        and all(total_counts[t] >= min_total_count for t in towers)
    )

def generate_balanced_problems(
    n_problems,
    towers,
    towers_per_problem=3,
    max_tries=100_000
):
    min_total_count = (n_problems * towers_per_problem) // len(towers)

    for _ in range(max_tries):
        problems = []
        used_combos = set()
        counts = Counter()

        for i in range(n_problems):
            possible = []

            for _ in range(500):
                combo = tuple(sorted(random.sample(towers, towers_per_problem)))

                if combo in used_combos:
                    continue

                test_counts = counts + Counter(combo)

                # Prefer underrepresented towers
                score = sum(test_counts[t] for t in combo)

                possible.append((score, combo))

            if not possible:
                break

            # choose one of the best-balanced options
            possible.sort(key=lambda x: x[0])
            combo = random.choice(possible[:20])[1]

            problems.append(combo)
            used_combos.add(combo)
            counts.update(combo)

        if len(problems) == n_problems and valid_schedule(problems, towers, towers_per_problem):
            return problems

    raise ValueError("Could not generate a valid balanced schedule.")

problems = generate_balanced_problems(20, towers)

for i, p in enumerate(problems, 1):
    print(f"Problem {i}: {p}")

print("\nCounts in first 6:")
print(Counter(t for p in problems[:6] for t in p))

print("\nCounts overall:")
print(Counter(t for p in problems for t in p))

problem_dict = {
    str(i + 1): list(problem)
    for i, problem in enumerate(problems)
}

print(problem_dict)
```
